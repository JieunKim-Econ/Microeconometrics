# 1. Introduction 
---

González (2013) examines the effects of a child subsidy on fertility, household expenditure, and maternal labor supply in Spain. The distinguishing feature of this paper is that it applies **regression discontinuity(RD) design** as well as **difference in differences (DID)** estimates. From an **“unanticipated” introduction** of a new law with benefit eligibility based on a certain date, the author exploits a natural experiment with randomization and the sharp cutoff. A universal child benefit, a one-time cash payment of €2,500, was announced on July 3 in 2007 in Spain, and mothers were eligible if their child was born after July 1 and this date was not announced beforehand. This setup enables RD analysis where the treatment effect is the difference in outcomes **between treated (who receive a benefit)** and **control families (who do not)** close to the cutoff.      


**Household expenditure and Labor Supply**\
For the expenditure and labor supply outcomes, the identification strategy is to compare the behavior of households who gave birth right before and right after the cutoff date. These families are statistically identical in all observable dimensions, except that some received the benefit and others did not. Hence, any “jump” at the cutoff date can be attributed to the policy, thus the identification strategy here is RDD. Furthermore, the author applies “difference-in-difference (DID)” estimates. To resolve the concerns of seasonal changes in the characteristics of women giving birth throughout the year, the author estimates specifications that contain multiple birth years and control for “seasonality” by including calendar month of birth fixed effects, thus supplementing RDD with DID estimates.  

**Fertility: Conception and Abortion**\
A tougher challenge as the author expresses is estimating the fertility effects. This is because there is no available dataset for the exact conception date. By using vital statistics data which include date of birth and weeks of gestation, the author constructs estimated conception date and finds a discontinuity in the number of conceptions at the cutoff date. Furthermore, the author sheds light on abortions with the claim that abortions should be discouraged if the benefit encourages fertility. The fertility analysis with estimated conception dates and abortions data also supplements RDD with DID specifications to resolve seasonality. 

**Main variables** 

| **Treatment**        | **Main outcomes**       | **Main Covariates** |
|---------------------  |-------------------------  |------------------------|
| 2,500EUR child benefit| Fertility                 | Age                    |
| .                     | Household Expenditure     | Educational attainment |                         
| .                     | Maternal Labor Supply     | Immigrant status       |
| .                     | Use in external childcare | Marital status         |


The purpose of this notebook is to revisit the main result and explore the policy impacts from more expended perspective. In brief, the author finds that the benefit significantly increased fertility. The result shows a positive jump in the number of conceptions right after the announcement date as well as a discrete drop in the number of abortions. For expenditure and maternal labor supply, families with the benefit did not increase consumption. Instead, eligible mothers left the work significantly longer after childbirth, leading to lower expenditure on external form of childcare.

##  2.1. Institutional Setting

**Introduction and Condition**__
On July 3, 2007, the Spanish president announced the introduction of a new universal child benefit. The one-time subsidy of €2,500 would be provided to all mothers giving birth on or after the announcement date. For practical reasons, the eligibility cutoff date was moved to July 1. The proposal became law in November and so called the first “baby-checks” were paid in late November 2007. The new benefit was universal without any income tests. The only requirement was the minimum 2 years of legal residence in Spain before giving birth. In 2008, the tax authorities reported payment of 491,557 “baby-checks”, equivalent with 95 percent of all births in Spain during the year. 

**The magnitude of the benefit**__
The amount of the benefit is considerable compared with monthly earnings. According to 2007 Wage Structure Survey, the monthly gross minimum wage for a full-time job in Spain was €570.60 in 2007, and about 20 percent of working women earned the minimum wage or below. Hence, the benefit was equal to 4.4 months of pay for a low-wage full-time worker.

**Explicit goals**__
The policy had two explicit goals. As stated in the law, the benefit was intended to help parents cope with the extra expenditures associated with childbirth as well as to encourage fertility, given the low prevalent birth rates and the population ageing trends. In addition, the law aims to facilitate the balancing of work and family, and to maintain the living standards of low-income households. 


##  2.2. Identification Strategy and Data 

The identification strategy is based on those following facts
*	The policy established a sharp cutoff in birth date for benefit eligibility
*	The cutoff was announced unexpectedly

There are important differences in the analysis of fertility versus household expenditures and labor supply:

* **A. Fertility**
* If the policy succeeded in encouraging fertility, we expect to observe a sudden increase in the number of conceptions right after the announcement date. Thus, the empirical approach is to _analyze the time series of conceptions over time and check a break around July of 2007_.

* **B. Household Expenditures and Maternal Labor Supply**
* The identification strategy is comparing the behavior of families who had a child right before and right after the cutoff. Thus, identification is achieved via _RDD, supplemented with DID_ specifications to account for potential month-of-birth effects.

The below is detailed discussion of the empirical specifications of two groups of outcomes. 

### 2.2.A. Fertility Effects

**Data: Vital Statistics for estimation of conception date**

The Spanish National Statistical Institutes provides micro-data on monthly births in Spain, which the author refers to as “vital statistics”. In particular, these data have information on weeks of gestation at birth. Thus, these vital statistics enable to estimate the date of conception with reasonable accuracy and analyze whether there was a discrete jump in the number of conceptions right after the introduction of the benefit. The analysis is performed at the monthly level since vital statistics data only reports the month of birth, not the exact date of birth.

**(1) Regression: Conceptions**

\begin{equation} 
(2) C_{m} = \alpha + \gamma_{1}m + \gamma_{2}(m * post) + \beta * post + \lambda  X_{m} + \epsilon_{m}
\end{equation}

*   $C$ : the natural log of the estimated number of conceptions in month $m$
*	$post$ : Binary indicator taking value 1 in all months starting July 2007 
*	$m$ : the month of conception, normalized to 0 for July 2007 (thus takes values −1 for June 2007, 1 for August 2007, etc.)
*	$X$ : the number of days in month $m$

The fundamental idea behind the equation is as follows; If the policy was successful, there would be an increase in the number of conceptions shortly after July 3, 2007. The identifying assumption is that no other factor affects conception rates discontinuously in July 2007. Then, coefficient $\beta$ of our main interest would capture a discrete jump in the monthly number of conceptions in July 2007, and it would be positive if the policy successfully encouraged fertility. 

**_Potential problems of (1)_**__
The author points out the above specification holds several potential problems. 
First, the fact that vital statistics do not report the exact date of birth hampers the ability to detect a discontinuity at the announcement date. More importantly, the broader the bins, the harder it becomes to rule out **"seasonality effects"**, which is systematic differences in conception rates by calendar month. In order to resolve this concern, the author includes calendar month of conception dummies, as shown in the next equation (2).

---
**(2) Regression: Conceptions with Seasonality**

\begin{equation} 
(2) C_{m} = \alpha + \gamma_{1}m + \gamma_{2}(m * post) + \beta * post + \lambda  X_{m} + \sum_{c=2}^{12}\mu_{c}\ month_{m} + \epsilon_{m}
\end{equation}



Now, coefficient $\beta$ would capture any discrete jump in the number of conceptions between June and July 2007, above and beyond the average difference in the number of conceptions between June and July of the surrounding years. This specification requires that the sample includes multiple years.

**_Remaining problems of (2) and introducing abortions_**__
Two concerns exist with (2). First, the month of conception is estimated with error. The author addresses this issue partially by using several alternative methods for estimating conception date. However, any remaining measurement error would bias coefficients toward zero. Secondly, we cannot expect conceptions to react immediately. The medical literature suggests that it typically takes three to six months for a healthy, fertile couple to conceive when actively trying. Hence, an increase in the number of couples trying to conceive would only result in actual conceptions gradually with some delay. 

The author addresses the last two concerns by analyzing the monthly incidence of abortions, instead of the estimated number of conceptions. This is feasible since the month of abortion is reported accurately, and the termination of a pregnancy can react immediately to the policy change. 

---
**(3) Regression: Combined effects on fertility; Conceptions and Abortions**

**Data of Abortions**__
Since the National Statistical Institute only reports the annual number of abortions, the author contacted the health authorities of 17 Spanish regions for monthly data. 12 of them, representing around 88 percent of total Spanish population in 2007, provided the data on monthly number of abortions between 2000 and 2009. The author estimates equations (1) and (2) using the natural log of number of abortions as the dependent variable.

**_The idea behind the final regression: Combined Effect_**__
Both an increase in the number of conceptions and a reduction in the number of abortions would result in a higher number of births after some time after the policy change. The author runs a final regression to capture this combined effect on fertility.

**1. Abortions**:  According to the author, the vast majority of abortions take place at less than 13 weeks of gestation. Then, a reduction in the incidence of abortions right after July 3 would lead to an increase in the number of births as early as January 2008. 

**2. Conceptions**:  On the other hand, the majority of pregnancies last 35 gestation weeks or more. Thus, any additional conceptions after July 3, 2007 would lead to an increase in the number of births in late February 2008, most likely in March. 

Therefore, the author expects the overall effect on births to appear from January 2008 but increase until September 2008, assuming that it may take up to six months to conceive. I drew the timeline to summarize the expected month of the policy effect appearance.

**Finally, the overall fertility effect** is illustrated as follows

\begin{equation} 
(3) B_{m} = \alpha + \gamma_{1}m + \gamma_{2}m^2 + \gamma_{3}m^3 + \beta * post + \sum_{c=2}^{12}\mu_{c}\ month_{m} + \epsilon_{m}
\end{equation}

Where $B$ is the natural log of the number of births in month $m$, and $post$ takes value 1 starting in January 2008. A cubic trend is included as a control, as well as calendar month dummies. 

Coefficient $\beta$ of our main interest captures the overall increase in the number of births after policy introduction. In order to capture any potential dynamics, the author includes a set of "$post$" dummies, grouping the post-benefit period in 6-months bins. The author expects the increase in births to appear between months 6 (January 2008) and 15 (September 2008).

### 2.2.B. Household Expenditures and Labor Supply

\begin{equation} 
(4) Y_{im} = \alpha + \gamma_{1}m + \gamma_{2}(m * post) + \beta * post + \Pi X'_{im} + \epsilon_{im}
\end{equation}

The sharp cutoff established for benefit eligibility leads to a regression discontinuity analysis, where we can compare the outcomes of households with a new child right before and right after the cutoff. The identifying assumption is that no other factor affected families with children born after June 30, 2007 discontinuously. 

Equation (4) is similar to equation (1), except the **individual-level subscript** $i$. $Y$ is an outcome variable (e.g. Total household expenditure or a measure of maternal labor supply) for household $i$, who had a child on month $m$. Month of birth $m$ (the running variable) is again normalized to zero at the threshold, July 2007. The vector $X$ includes household-level controls, and $\epsilon$ is the residual. The parameter of our main interest, $\beta$, captures any potential discontinuity or "jump" in $Y$ at the cutoff.

**With this regression, the author identifies total three effects as below**
*	Changes in household expenditure patterns
*	The potential labor supply effects
*	Changes in childcare use

In the main specification, $Y$ is observed in 2008, or about 12 months after birth for children born at the cutoff date. **The main sample includes households with children born between 9 months before and 9 months after the policy change (i.e. between October 2006 and March 2008)**, so that the children are between zero and two years old when observed. In additional specifications, the author varies the number of months around the threshold included in the sample, up to only two months on either side of the cutoff. Each set of dependent variables use different datasets. The expenditure analysis uses Household Budget Survey (HBS) data, while the main labor supply analysis uses data from the larger Labor Force Survey(LFS). 


**_Validity check of RDD_**__
In order to confirm the validity of the RDD approach, two checks should be performed (Lee and Lemieux 2010); **No discontinuity** and **household characteristics around the threshold**.

* First, there should be **_no discontinuity in the number of births around the threshold._** Since date of birth determined benefit eligibility, the policy produced an incentive to postpone birth to after the cutoff date. However, such shifting is highly unlikely as the benefit was announced three days after the threshold date. With vital statistics, the author shows that there were no discontinuity in the number of births around July 1, 2007. The results are reported in <a href="#3.2.1.-Validity-Check">3.2.1. Validity Check</a>

* Second, we should check the balance in household characteristics around the threshold. If the treatment is “as if” randomly assigned, there should be **_no significant differences between treated and control families_**. Thus, the author estimates regressions such as (4) where the dependent variables are a range of household characteristics (age, education level of the parents, immigrant status, etc). We will see the results in <a href="#3.2.1.-Validity-Check">3.2.1. Validity Check</a>
    
**_Resolving Seasonality with DID_**__
Since the running variable is discrete month of birth, there is a concern that an RDD analysis could capture seasonality effects, i.e., any potential systematic differences between June and July births that are unrelated to the benefit. The author addresses this concern by additionally running difference-in-difference specifications, by including multiple birth-years and calendar month of birth fixed effects. Thus, a discontinuity between June and July 2007 births would only be interpreted as a treatment effect if it was larger than the average June-July difference in other surrounding years.

Two different DID exercises are implemented. Frist, all households are interviewed in 2008, and the sample consists of families with children born between 2005 and 2008 (both inclusive). Thus, all families are observed at the same point in time, and the ages of children are between zero and three. Next, the second DID specification also includes observations from families with children born between 2005 and 2008, but now the author merges survey data from 2006 to 2009, so that families are always interviewed the year following childbirth (in 2008 for 2007 births, in 2009 for 2008 births, etc.). Thus, in the second DID, the children are observed at almost the same age, but in different calendar years.

# 5. Conclusion
---


In general, I find the results reported in this notebook consistent with González's main results. With available data and STATA code, the estimation of the RDD regression supplemented by DID specification are precisely reproduced, supporting the author’s main findings. However, the absence of HBS 2009 data and LFS 2009 data is an obstacle to reproduce the estimates of the last column DID(8) in Table 4, 5, 6 and the bottom panel of Figure 4 in the paper. As the author highlights DID estimation to address the seasonality effect, only one DID analysis out of two limits a chance to have comprehensive interpretation of the results. Aside from the replication, by embracing father’s characteristics, I extend the analysis of the labor supply effect with different subpopulation grouped by the education level of parents and marital status.

González analyzes the effects of a €2,500 universal child benefit introduced in Spain in 2007. 
The distinguishing feature of the paper is that the author evaluates the policy impact on a broad range of outcomes, showing the interactions between mother’s (economic) choices: fertility, household expenditure, and the labor supply. The evidence suggests that the child policy succeeded in achieving its explicit goal of increasing fertility by around 6 percent, derived from the combined effect of an increase in conception and the reduced incident of abortions. On the other hand, the study reveals no significant change in household total expenditure and child-related expenditure. Instead, the estimates of the labor supply and expenditure of use in private childcare show that mothers rather decided to “buy time” and stay at home with their newborn child by reducing use in external childcare and delaying their return to workplace during one year after the birth. 

For countries with low prevalent birth rates and the population ageing trends, the paper would give valuable insights into policy reform for demographic change and labor market structure. While the study gives the comprehensive view of the impact of the child grant but over two years at most, the study for long-term consequences would make a richer contribution to exploring the lasting effect of expansion in maternal leave on child development as well as household expenditure patterns.
